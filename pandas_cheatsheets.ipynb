### Фильтрация по квантилю (анализ выбросов)

```python
# выбираем данные до квантиля по столбцу city_development_index
q_025 = df['city_development_index'].quantile(0.25)
df[df['city_development_index'] < q_025]
```

```python
# между двумя квантилями

q_025 = df['city_development_index'].quantile(0.25)
q_05 = df['city_development_index'].quantile(0.5)
df[df['city_development_index'].between(q_025, q_05)]['city_development_index'].hist()
```

```python
# средний % теэ, кто хочет сменить работу
df['target'].mean()
```

```python
# распределение % с учетом пола
df.groupby('gender')['target'].agg(['mean', 'count'])
```

```python
# разделяем 'city_development_index' на 5 равных групп, потому что по дефолту не читаемо
df['city_development_index_group'] = pd.cut(df['city_development_index'], 5)
```

```python
df.groupby('city_development_index_group', dropna = False)['target'].agg(['count', 'mean'])
```

```python
# разделяем на 5 равных групп с удалением дубликатов
df['city_development_index_group_q'] = pd.qcut(df['city_development_index'], 5, duplicates = 'drop')
df.groupby('city_development_index_group_q', dropna = False)['target'].agg(['count', 'mean'])
# строим распределение тех, кто хочет сменить работу, в зависимости от city_development_index
t = df.groupby('city_development_index_group_q', dropna = False)['target'].agg('mean')
t.plot()
# вывод - чем больше развитость города, тем меньше % тех, кто хочет сменить работу
```

```python
# разбивка на интересующее нас количество интервалов
# в нашем случае бессмысленно, но для потенциально контроля может быть полезно. float('inf') - бесконечность
df['city_development_index_group_custom'] = pd.cut(df['city_development_index'], 
                                                    [-float('inf'), 0, 0.1, 0.2, 0.5, float('inf')])
custom = df.groupby('city_development_index_group_custom', dropna = False)['target'].agg(['count', 'mean'])
custom.plot()
```
