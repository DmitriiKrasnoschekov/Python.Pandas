```python
# разделяем 'city_development_index' на 5 равных групп, потому что по дефолту не читаемо
df['city_development_index_group'] = pd.cut(df['city_development_index'], 5)
```

```python
# разделяем на 5 равных групп с удалением дубликатов
df['city_development_index_group_q'] = pd.qcut(df['city_development_index'], 5, duplicates = 'drop')
df.groupby('city_development_index_group_q', dropna = False)['target'].agg(['count', 'mean'])
# строим распределение тех, кто хочет сменить работу, в зависимости от city_development_index
t = df.groupby('city_development_index_group_q', dropna = False)['target'].agg('mean')
t.plot()
# вывод - чем больше развитость города, тем меньше % тех, кто хочет сменить работу
```

```python
# разбивка на интересующее нас количество интервалов
# в нашем случае бессмысленно, но для потенциально контроля может быть полезно. float('inf') - бесконечность
df['city_development_index_group_custom'] = pd.cut(df['city_development_index'], 
                                                    [-float('inf'), 0, 0.1, 0.2, 0.5, float('inf')])
custom = df.groupby('city_development_index_group_custom', dropna = False)['target'].agg(['count', 'mean'])
custom.plot()
```

Группировка категориальных переменных

```python
df.groupby('education_level', dropna=False)['target'].agg(['count', 'mean'])
```

```python
# сгруппируем уровни образования    
def education_group(x):
    if x in ['High School', 'Primary School']:
        return 'School'
    elif x in ['Masters', 'Phd']:
        return 'Masters_and_phd'
    else:
        return x
```

```python
# применяем функцию к столбцу и сохраняем в новый столбец
df['new_education_level'] = df['education_level'].apply(education_group)
```

```python
# делаем перегруппировку с новой группой
df.groupby('new_education_level', dropna=False)['target'].agg(['count', 'mean'])
```

```python
# Функции с несколькими столбцами
# Прибавляем training_hours для людей со стпенью phd + 1000

def new_training_hours(row):
    if row['education_level'] == 'Phd':
        return row['training_hours'] + 1000
    else:
        return row['training_hours']
```

```python
# применяем функцию
df['new_training_hours'] = df.apply(new_training_hours, axis = 1)
# фильтруем записи с education_level == Phd
df[df['education_level'] == 'Phd']
```

```python
# Функцию Pivot Table как аналог groupby

df.pivot_table(index = 'new_education_level', values = 'target', aggfunc=['count', 'mean'])

# как мы видим, NaN нет, функция dropna в этом случае далает другое. 
```

```python
# Чтобы подтянуть Nan значения, нужно собственноручно их заполнять. Пример:

df['new_education_level'] = df['new_education_level'].fillna('no_info')
df.pivot_table(index = 'new_education_level', values = 'target', aggfunc=['count', 'mean'])

# Помимо этого, есть мульти индекс, что не так хорошо.
```
