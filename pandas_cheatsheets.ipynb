Фильтрация и Ворнинги

```python
# отфильтровываем записи со значением city_40 в столбце city
df[df['city'] == 'city_40']
```

```python
# аналогичная операция через df.query
df.query('city == "city_40"')
```

```python
# фильтрация по нескольким пунктам
df[(df['city'] == 'city_40') & (df['gender'] == 'Male') & (df['training_hours'] > 80)]
```

```python
# запись отфильтрованных строк в новые данные. Без этого будет ворнинг
df_city_gender = df[(df['city'] == 'city_40') & (df['gender'] == 'Male') & (df['training_hours'] > 80)].copy()
df_city_gender['new'] = 'new'
df_city_gender
```
